```python
# single level inheritance :(single parent and child)
class company:

    staff=100

    def details(self):
        print("Top most compant")
        
class BNT(company):
    student=150

company1=BNT()
print(company1.staff)
print(company1.student)
company1.details()
```

```python
class animals:


    def eat(self):
        self.animal="dog"
        self.food="biryani"
        print(self.animal,self.food)
        

class dog (animals):

    def sound(self):
        self.noise="Bark"
        print(self.noise)

a=dog()
a.sound()
a.eat()
```

```python
# Hierarchical Inheritance :("single parent and multiple child class")

class company: # parent class
    
    staff=100

    def about(self):
        print("Top most company")

class FITA(company): # child class 1
    students=200

class wipro(company): #child class 2
    students=300

company1=FITA
company2=wipro

print(company1.students)
print(company2.students)
```

```python
# Hierarchical Inheritance inhertitance example

class job:
    role="Advance python"
    
    def detail(self):
        print("Employee  details")
        
class emp1(job):
    name="rio"
    age=22
    salary=100000
    
class emp2(job):
    name="santoo"
    age=21
    salary=80000

e1=emp1
e2=emp2


print("Employee name:",e1.name,"AGE:",e1.age,"salary:",e1.salary , e1.role)
print("Employee name:",e2.name,"AGE:",e2.age,"salary:",e2.salary, e2.role)
```

```python
# multi level inheritance (grandfather - father -son)

class company:
    staff=20
    
class fita(company):
    students=100
    course=6
    
class bnt(fita):
    students=200
    
company1=bnt()
print(company1.students)
print(company1.course)
```

```python
# multiple inheritance (one child and multiple parent)

class father:
    hair="white"
    
class mother:
    hair="grey"
    skin="brown"
    
class child(mother,father):
    legs=2
    hands=2
    
child=child()

print(child.hair)
print(child.skin)
```
