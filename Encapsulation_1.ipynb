```python
# public variable in encaplusation
class student:
    def __init__(self):
        self.name="sam"
s=student()
print(s.name)
```

```python
# protection variable in encaplusation
class emp:
    def __init__(self):
        self._salary=200000
e=emp()
print(e._salary)
```

```python
# private variable in encaplusation
class bank:
    def __init__(self):
        self.__balance=2345678
        
    def show(self):
        print(self.__balance)
        
        
b=bank()
b.show()
```

```python
class student:
    def __init__(self):
        self.__mark=30
        
    def setmarks(self,m):
        self.__marks=m
        
    def getmarks(self):
        return self.__marks
        
s=student()
s.setmarks(500)

print(s.getmarks())
```

```python
class ATM:
    def __init__(self):
        self.__pin=2192
        
    def setpin(self,p):
        self.__pin=p
        
    def getpin(self):
        return self.__pin
    
a = ATM()
a.setpin(1414)
print(a.getpin())
```

```python
class emp():
    def __init__(self):
        self.__salary=20000
        
    def setsal(self,s):
        self.__salary=s
        
    def getsal(self):
        return self.__salary
    
e=emp()
e.setsal(50000)

print(e.getsal())
```

```python
#create class car private variable __speed getter setter method

class car:
    
    def __init__(self):
        self.__speed=30
    
    def setsp(self,s):
        self.__speed=s
        
    def getsp(self):
        return self.__speed
c=car()
c.setsp(70)
print(c.getsp())
```

```python
# create class book protector variable _pages

class book:
    def __init__(self):
        self._pages=20
        
    
b=book()
print(b._pages)
```

```python
# login system using encapsulation

class login:
    def __init__(self):
        self.__username="santoo"
        self.__pw=123
        
    def setlogin(self,new_user , new_pw):
        self.__username=new_user
        self.__pw=new_pw
        
    def getlogin(self):
        user = input("Enter username :")
        pwd = int(input("Enter a password:"))
        
        if user==self.__username and pwd == self.__pw :
            print ( "login sucessfully")   
        else:
            print("invalid")
            
        
l=login()
l.setlogin("sam",321)

l.getlogin()
        
        
        
    
    
```
