```python
grid=[
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9]
]

def safe(row,col,num):
    
    for x in range(9):
        if grid [row][x]==num:
            return False
        
    for x in range(9):
        if grid [x][col]==num:
            return False
        
    start_row = row-row%3
    start_col=col-col%3
        
    for i in range(3):
        for j in range(3):
            if grid[start_row +i] [start_col +j] == num:
                return False
            
    return True

def solve():
    for row in range(9):
        for col in range(9):
            
            if grid[row][col]==0:
                for num in range(1,10):
                    
                    if safe(row , col , num):
                        grid[row][col]=num
                        
                        if solve():
                            return True
                        
                        grid[row][col]=0
                        
                return False
    return True

solve()

for row in grid:
    print(row)
    


                
                            
    
                
```
