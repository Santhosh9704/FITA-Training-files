```python
from abc import ABC, abstractmethod

class animal(ABC):

    @abstractmethod
    def sound(self):
        pass


class dog(animal):

    def sound(self):
        print("Dog bark")


d = dog()
d.sound()
```

```python
from abc import ABC,abstractmethod
class  vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
class car (vehicle):
    def start(self):
        print("car started")
        
c=car()

c.start()
```

```python
# create a abstract class shape , abstractmethod area ,child class square print area * abstractmethod
from abc import ABC , abstractmethod
class shape(ABC):
    @abstractmethod
    def area(self):
        pass
class square(shape):
    def area(self , a):
        print ( " Area: ",a*a)
        

s=square()
s.area(30)
    
```

```python
# craete abstrct empolyee class , abstractmethod salary class ,child classs manager , print manager salary

from abc import ABC,abstractmethod
class  emp(ABC):
    @abstractmethod
    def salary(self):
        pass
class manager (emp):
    def salary(self):
        print(450000)
        
m=manager()

m.salary()
```

```python
# craete abstarct notification , abstractmethod send() , child class email(), print email notification send

from abc import ABC,abstractmethod
class  notification(ABC):
    @abstractmethod
    def send(self):
        pass
class email (notification):
    def send(self):
        print("Email notification send")
        
e=email()

e.send()
```
