```python
# generator example
def generator():
    yield 1
    yield 2
    yield 3
    
g = generator()
print(next(g))
print(next(g))
print(next(g))
```

```python
# count the number using generator
def count():

    for i in range(1, 6):
        yield i

g = count()
print(next(g))
```

```python
def number():
    yield 10
    yield 20
    yield 30
    
g = number()
for i in g:
     print(i)
```

```python
def student ():
    roll = 1
    
    while roll <= 5:
        yield roll
        roll += 1
for i in student():
    print(i)
```

```python
#odd or even number using generator
def oddeven(n):
   
    for n in range(1,n+1):
        
        if n%2==0:
            yield n,'EVEN'
        else:
            yield n,"ODD"
        n += 1
n=int(input("Enter the number :"))       
g = oddeven(n)

for i in g:
    print(i)
```

```python
# reverse a string using generator
def reverse(s):
    rev=""
    for ch in s:
        rev=ch+rev
    yield rev
s=input("Enter the string :")
g=reverse(s)
print(next(g))
```

```python
#reverse the number using generator
def num(n):
    sum=0
    while n>0:
        digit=n%10
        sum=sum*10+digit
        n=n//10
    yield sum
n=int(input("Enter the number :"))
g=num(n)
print(next(g))
```

```python
#remove duplicate using generator
def r_duplicate(l):
    dup=[]
    for i in l :
        if i not in dup:
            dup.append(i)
            
    yield dup

l=[1,2,3,4,5,1,2,3]
g=r_duplicate(l)
for i in g:
    print(i)
```

```python
def prime(n):
    for i in range(2,n+1):
        for j in range(2,i):
            if i%j==0:
                break
        else:
            yield i
n=10
g=prime(n)
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```
