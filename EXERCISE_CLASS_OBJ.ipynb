```python
#student details
class student:
    
    def __init__ (self,name,age):
        self.name=name
        self.age=age
        
s1=student("sam",22)

print(s1.name)
print(s1.age)
```

```python
#2 .car details
class CAR:
    def __init__ (self,brand,prize):
        
        self.brand=brand
        self.prize=prize
        
c1=CAR("BMW", 2000000)

print("Car model : ", c1.brand)
print("prize: ", c1.prize) 
```

```python
#3 . book details
class book:
    
    def __init__ (self,title,author):
        
        self.title=title
        self.author=author
        
b=book("PYTHON","Stephen")

print("BOOK TITLE : ",b.title)
print("Author : ",b.author)
```

```python
#6. bank acccount info
class bank_account:
    def __init__(self, account_number, balance):
        
        self.account_number = account_number
        self.balance = balance
        
p1 = bank_account(63212678, 40000)

print("Account number:", p1.account_number)
print("Balance amount:", p1.balance)
```
