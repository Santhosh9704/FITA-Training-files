```python
# polymorsphism 1st type - mmethod overloading
class Math:

    def add(self, a, b=0, c=0):
        print(a + b + c)


m = Math()

m.add(10, 20)
m.add(10, 20, 30)
```

```python
# polymorphism 2nd type - method overriiding
class Animal:

    def sound(self):
        print("Animal sound")


class Dog(Animal):

    def sound(self):
        print("Dog barks")

class cat (Animal):
    
    def sound(self):
        print("cat meow meow")


d = Dog()
a=Animal()
c=cat()
a.sound()
d.sound()
c.sound()
```

```python
# polymorphism 2nd type - method overriiding
class father:
    def __init__(self):
        print("father  constructor")
    def say(self):
        print("HEllo father")
        
class child(father):
    def say(self):
        print("hello child")
        
child=child()

child.say()

f=father()
f.say()
```
